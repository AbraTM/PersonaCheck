## Dataset (from Kaggle)
#### Features seen in extroversion and introversion

```python
df = pd.read_csv('../data/personality_dataset.csv')
df.head(10)
```

### Cleaning / Pre - Processing the data
#### Removing invalid numbers (NaN)

```python
df.dropna(inplace = True)
df.reset_index(drop = True, inplace = True)
```

## Data is now properly processed and can be fitted into a NAIVE BAYES model for prediction
### Using Scikit-learn's NAIVE BAYES model implementation

#### Seperating the Input and Output Data

```python
# Input data / Features
X = encoded_df.drop(["Personality"], axis=1)
y = encoded_df["Personality"]
```

#### Then, splitting the data into train, test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

#### Fitting the data

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
gnb.score(X_train, y_train)
```

```python
gnb.score(X_test, y_test)
```
